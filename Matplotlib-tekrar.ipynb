## Grafik Olu≈üturma

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
y=[1,4,9,16]
#plt.plot(x,y,color="red",linewidth="5") #grafik cizgi rengi ve kalƒ±nlƒ±ƒüƒ±
#plt.plot(x,y,"--g")
#marker line color
plt.plot(x,y,"o--r")

plt.axis([0,6,0,20])
#x ekseninin min degeri 0 max degeri6 ,y ekseninin min degeri 0 max degeri 20 olur 
plt.title("Grafik Ba≈ülƒ±ƒüƒ±") #grafik baslƒ±k
plt.xlabel("X Label") #x eksenƒ± baslƒ±k
plt.ylabel("Y Label") #y eksenƒ± baslƒ±k
plt.show()
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label="linear",color="red")
plt.plot(x,x**2,label="quadratic",color="yellow")
plt.plot(x,x**3,label="cubic",color="blue")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple plot")
plt.legend() # cizgilerin nelere karsƒ±lƒ±k geldƒ±gƒ±nƒ± sol ust kosede ƒ±fade eder
plt.show()
```

```python
#grafƒ±k cƒ±zƒ±len alana axes denir
#bƒ±r alana bƒ±rden fazla grafƒ±k cƒ±zebƒ±lƒ±rƒ±z ,plt.subplot(2) metotunda 2 parametresƒ±nƒ± verƒ±rsek grafƒ±k cƒ±zmek ƒ±cƒ±n 2 axes alanƒ± olusturmus oluruz 
# bƒ±ze bƒ±r fƒ±gure donecek
x=np.linspace(0,2,100)
fig,axes=plt.subplots(3)
#her bƒ±r axes e erƒ±smek ƒ±cƒ±n
axes[0].plot(x,x,color="red")
axes[0].set_title("linear")

axes[1].plot(x,x**2,color="green")
axes[1].set_title("quadratic")

axes[2].plot(x,x**3,color="yellow")
axes[2].set_title("cubic")
plt.tight_layout()#baslƒ±klarƒ±n yerlerƒ±nƒ± duzelttƒ± 


plt.show()
```

```python
x=np.linspace(0,2,100)
fig,axes=plt.subplots(2,2)
fig.suptitle("grafik baslƒ±gƒ±")
axes[0,0].plot(x,x,color="red")
axes[0,1].plot(x,x**2,color="green")
axes[1,0].plot(x,x**3,color="yellow")
axes[1,1].plot(x,x**4,color="blue")
plt.show()
```

```python
x=np.linspace(-10,9,20)
y=x**3
z=x**2
figure=plt.figure()
#axes=figure.add_axes([0.2,0.2,0.8,0.8])
#soldan ve sagtan 0.2 uzaklƒ±kta genƒ±slƒ±k ve yukseklƒ±k olarak 0.8 (sayfanƒ±n tamamƒ± 1 dir)
axes_cube=figure.add_axes([0.1,0.1,0.8,0.8])
axes_cube.plot(x,y,"b")
axes_cube.set_xlabel("x axis")
axes_cube.set_ylabel("y axis")
axes_cube.set_title("cube")

axes_square=figure.add_axes([0.2,0.65,0.2,0.2])
axes_square.plot(x,z,"r")
axes_square.set_xlabel("x axis")
axes_square.set_ylabel("y axis")
axes_square.set_title("square")

plt.show()
```

```python
 
x=np.linspace(-10,9,20)
y=x**3
z=x**2
figure=plt.figure()
axes=figure.add_axes([0.0,0.0,1,1])

axes.plot(x,z,label="Square")
axes.plot(x,y,label="Cube")
axes.legend(loc=1)
plt.show()
```

```python
x=np.linspace(-10,9,20)
y=x**3
z=x**2
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,8))
#figurun boyutunu da ayarlayabƒ±lƒ±rƒ±z fƒ±gsƒ±ze ile 
axes[0].plot(x,y)
axes[0].set_title("Cube")

axes[1].plot(x,z)
axes[1].set_title("Square")

plt.tight_layout()
fig.savefig("figure1.png")
#figuru kaydetmek ƒ±cƒ±n

plt.show()
```

```python
import matplotlib.pyplot as plt
#grafik t√ºrleri
#1)stack plot
yil=[2011,2012,2013,2014,2015]
oyuncu1=[8,10,12,7,9]
oyuncu2=[7,12,5,15,21]
oyuncu3=[18,20,22,25,19]

plt.plot([],[],color="y",label="oyuncu1")
plt.plot([],[],color="r",label="oyuncu2")
plt.plot([],[],color="b",label="oyuncu3")
plt.stackplot(yil,,oyuncu1,oyuncu2,oyuncu3,colors=["y","r","b"])
plt.title("Yƒ±llara gore atƒ±lan gol sayƒ±sƒ±")
plt.xlabel("yil")
plt.ylabel("goller")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Veriler
yil = [2011, 2012, 2013, 2014, 2015]
oyuncu1 = [8, 10, 12, 7, 9]
oyuncu2 = [7, 12, 5, 15, 21]
oyuncu3 = [18, 20, 22, 25, 19]

# Grafik olu≈üturma
plt.stackplot(yil, oyuncu1, oyuncu2, oyuncu3, labels=["Oyuncu 1", "Oyuncu 2", "Oyuncu 3"], colors=["y", "r", "b"])
plt.title("Yƒ±llara G√∂re Atƒ±lan Gol Sayƒ±sƒ±")
plt.xlabel("Yƒ±l")
plt.ylabel("Goller")
plt.legend()
plt.show()
```

```python
#3) bar grafƒ±gƒ±
import matplotlib.pyplot as plt
plt.bar([0.25,1.25,2.25,3.25,4.25],[50,40,70,80,20],label="bmw",width=.5)
plt.bar([0.75,1.75,2.75,3.75,4.75],[80,20,20,50,60],label="audi",width=.5)
plt.legend()
plt.xlabel("gun")
plt.ylabel("mesafe(km)")
plt.title("arac bilgisi")
plt.show()
```
