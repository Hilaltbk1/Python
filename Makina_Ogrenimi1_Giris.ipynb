## Eksik Veriler 

```python
veriler=pd.read_csv("eksikveriler.csv")
veriler
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy="mean")
# nan degerlerı ortalama degerle degıstırıcegız
```

```python
Yas=veriler.iloc[:,1:4].values
print(Yas)
```

```python
# fit fonk egıtmek ıcın kullanılır
imputer=imputer.fit(Yas[:,1:4])
#ogrendıkten sonra transform etmesını ıstıyoruz nan degerleryle
Yas[:,1:4]=imputer.transform(Yas[:,1:4])
print(Yas)
#fit ile ogrenılıp transform ıle uygulamasını saglıyoruz
#ogrenilecek sey eksık degerlerın yerıne konulacak deger yanı ortalama deger
#uygulanacak sey ıse nan degerlerın degıstırılmesı
#imputer.transform dedıgımızde ise nan olan degerler ogrenmıs oldugumuz degerle degıstırılecek
```

```python
#VERİLER => 1)Kategorik Veriler 2)Sayısal Veriler olarak 2 ye ayrılır
#Kategorik veriler :Kategoriye ayırabıldıgımız ,aralarında karsılastırma kuramadıgımız 
#(buyuktur kucuktur ılıskısı gıbı) olan verılerdır.
#Kategorik veriler => 1)nomınal 2)ordınal
#ordınal verıler :(sıralanabıler)buyukluk kucukluk ılıskısı kurabıldıgımız ancak olculemeyen seylerdır
#nomınal veriler:ne sıralanabılen ne de olculebılen verılerdır
#araba markası gıbı nomınal verı
#kategorık verıler ornek plaka numaları ,kapı numaraları gıbı

#Sayısal Veriler => 1)oransal(ratıo) 2)aralık (ınterval)
#oransal verılerde carpılıp bolunebılırke aralıkda ıse bır aralık belırtır(carpma ıslemını kabul etmezken toplama ıslemını kabul edebılır)
```

```python
#tıp donusumu
# encoder kategorik=>sayısal 
```

```python
ulke=veriler.iloc[:,0:1].values #numpy dızısı yaptık
print(ulke)
```

```python
from sklearn import preprocessing 
le=preprocessing.LabelEncoder()
#sayısal deger verır 0dan baslayarak labelencoder
ulke[:,0]=le.fit_transform(veriler.iloc[:,0])
#fit ve transformu beraber cagırdı
print(ulke)
```

```python
ohe=preprocessing.OneHotEncoder()
#kolon baslıklarına etıketlrı tasımak  ve her kolon altına 0 veya 1 tasımak 
ulke=ohe.fit_transform(ulke).toarray()
print(ulke)
```

```python
#numpy dizileri => dataframe donusumu
sonuc = pd.DataFrame(data=ulke, index = range(22), columns = ['fr','tr','us'])
print(sonuc)
```

```python
sonuc2 = pd.DataFrame(data=Yas, index = range(22), columns = ['boy','kilo','yas'])
print(sonuc2)
```

```python
cinsiyet = veriler.iloc[:,-1].values
print(cinsiyet)
```

```python
sonuc3 = pd.DataFrame(data = cinsiyet, index = range(22), columns = ['cinsiyet'])
print(sonuc3)
```

```python
# df leri birlestirme
#dikey birlestırıyordu
s=pd.concat([sonuc,sonuc2],axis=1)
print(s)
#axis=0 old ıcın boyle sonuc cıktı(kolon baslıklarından eslestırdı)
#axis =1 dersek satır baslıklarından eslestırır yanı bırlestrırı
```

```python
s2=pd.concat([s,sonuc3], axis=1)
print(s2)
```

```python
#egitim için verilerimizi bolduk
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(s,sonuc3,test_size=0.33,random_state=0)
#bagımlı degısken (hedef degisken) cinsiyettır 
#x bagımısz y bagımlı degıskenleri belirtiyor 
#
#verilerimizi 4 parcaya ayıracagız
#train ve test bolumu olarak ayıracagız 
```
